# Сокращяем код

```python
# Алгоритм для 8 забегов
# Сократим для проведения эксперимента
from random import shuffle # для перемешивания (создания случайностей выбора)
import numpy as np # матрицу чтоб создать
import pandas as pd # будем датаффрейм делать
# Для результатов экмперементов
df8 = pd.DataFrame(columns = ['first', 'second',  'third'])
# 100 экспериментов
for i in range(100):
    # создаем табун и распределяем по 5 групам
    x =[i for i in range(1,26)]
    shuffle(x)
    shuffle(x)
    tabun = np.array(x)
    tabun.shape = 5,5
    finish = np.sort(tabun, axis = 1)
    # участники забега
    zabeg = finish[:,-1]
    # в сторонке
    remains =  np.delete(finish, [4], axis = 1)
    finish6 = np.sort(zabeg)  
    first = finish6[-1]
    # номер пятерки победителя
    nm = np.argmax(zabeg)
    # Следующий за победителем
    next1 = remains[nm, -1]
    # ставим в забег
    zabeg[nm] = next1
    # Удаляем из сторонки
    remains[nm, -1] = 0
    remains = np.sort(remains, axis = 1)  
    finish7 = np.sort(zabeg)  
    second = finish7[-1]
    # номер пятерки победителя
    nm = np.argmax(zabeg)
    # Следующий за победителем
    next1 = remains[nm, -1]
    # ставим в забег
    zabeg[nm] = next1
    remains[nm, -1] = 0
    remains = np.sort(remains, axis = 1)  
    finish8 = np.sort(zabeg)  
    third = finish8[-1]
    sample = [{'first':first, 'second':second,  'third':third}]
    df8 = df8.append(sample, ignore_index=True)

```

```python
# Алгоритм для 7 забегов
# Потом забег среди победителей, так мы находит 1го самого быстрого. 
# И еще один забег из двух верблюдов (берем второго и третего) в пятерке, 
# где находится лидер всех верблюдов. Еще двоих берем из пятерки, 
# где лидер пятерки пришел вторым в забеге лидеров (берем первого и второго). 
# И последнего верблюда (лидера в рассматриваемой пятерки) берем из пятерки, 
# где лидер пришел третьим в забеге лидеров.
# Так мы найдем второго и третьего верблюда
# Сократим для проведения эксперимента
from random import shuffle # для перемешивания (создания случайностей выбора)
import numpy as np # матрицу чтоб создать
import pandas as pd # будем датаффрейм делать
# Для результатов экмперементов
df7_1 = pd.DataFrame(columns = ['first', 'second',  'third'])
# 5000 экспериментов
for i in range(50000):
    # создаем табун и распределяем по 5 групам
    x =[i for i in range(1,26)]
    shuffle(x)
    shuffle(x)
    tabun = np.array(x)
    tabun.shape = 5,5
    
    finish = np.sort(tabun, axis = 1)
    # участники забега
    zabeg = finish[:,-1]
    # в сторонке
    remains =  np.delete(finish, [4], axis = 1)
    finish6 = np.sort(zabeg)  
    first = finish6[-1]

    # номер пятерки победителя
    nm = np.argmax(zabeg)
    zabeg7 = []
    #(берем второго и третего) в пятерке, где находится лидер всех верблюдов.
    zabeg7.append(remains[nm, -1])
    zabeg7.append(remains[nm, -2])
    # Еще двоих берем из пятерки, где лидер пятерки пришел вторым в забеге лидеров (берем первого и второго)
    # пятерки второго пришедшего в 6м забеге
    # викидываем лидера
    zabeg[nm] = 0
    # пятерка второго пришедшего в 6м забеге
    nm2 = np.argmax(zabeg)
    
    zabeg7.append(remains[nm2, -1])
    zabeg7.append(finish6[-2]) # это второй в 6м забеге
    
    # И последнего верблюда (лидера в рассматриваемой пятерки) берем из пятерки, 
    # где лидер пришел третьим в забеге лидеров.
    zabeg7.append(finish6[-3])

    finish7_1 = np.sort(zabeg7) 
    second = finish7_1[-1]
    third = finish7_1[-2]
   
    sample = [{'first':first, 'second':second,  'third':third}]
    df7_1 = df7_1.append(sample, ignore_index=True)
```
